```
iris = datasets.load_iris()
X = iris.data[:, :]  # we only take the first two features.
y = iris.target

#there are 3 labels, lets reduce the problem and define only 2 labels : the label 1 remains 1, and label 2 and 3 are merged.
y[y>=1] = 1

print(X.shape)
print(y.shape)
print(y)

plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1,edgecolor='k')
```

**Exercise 4 :** Implement a logistic regression model that classify the iris into the right class. (Use the Gradient Descent algorithm to find the parameters)

```
theta0=2
theta1=2
n=0.01 # attention: ici n est le taux d'apprentissage et non le nombre d'itérations donc on doit mettre un nombre très petit
functionlogistic=np.exp(theta0+theta1*X[:,1])/(1+np.exp(theta0+theta1*X[:,1]))
#cohérence: numérateur /dénominateur  avec X[:,1]
gradienttheta0=-np.sum(y- 1/(1+np.exp(-theta0-theta1*X[:,1])))
#gradienttheta0=-np.sum(y- functionlogistic)
gradienttheta1=-np.sum(X[:,1]*y- 1/(1+np.exp(-theta0-theta1*X[:,1])))
#gradienttheta1=-np.sum(X[:,1]*y- functionlogistic)



while gradienttheta0 > 0.01 and gradienttheta1 > 0.01:
  #Attention: mettre les deux gradients dans la même boucle while car dans la formule vue en cours, les variables theta0 et theta1 sont dépendantes
  
  
  theta0=theta0-n*gradienttheta0
  gradienttheta0=-np.sum(y-1/(1+np.exp(-theta0-theta1*X[:,1])))
  #attention: theta0= theta0- n*gradient ne pas oublier le gradient
  theta1=theta1-n*gradienttheta1
  #attention: theta1= theta1- n*gradient ne pas oublier le gradient
  gradienttheta1=-np.sum(X[:,1]*y- 1/(1+np.exp(-theta0-theta1*X[:,1])))
  
print('Theta0: ', theta0)
print('Theta1: ', theta1)
print('Gradient pour theta0',gradienttheta0)
print('Gradient pour theta1',gradienttheta1)


#calculer la précision
iterat=0
for i in range(len(X)):
  if functionlogistic.all()<0.5:
    y[i]=0
  else:
    y[i]=1
    iterat=iterat+1
    

print("Nombre de y=1: ",iterat)
print("La précision est égale à:",(iterat/len(X))*100)
```

**Exercise 5 :** Use now the Stochastic Gradient Descent algorithm

```
theta0=2
theta1=2
n=0.01 # attention: ici n est le taux d'apprentissage et non le nombre d'itérations donc on doit mettre un nombre très petit
functionlogistic=np.exp(theta0+theta1*X[:,1])/(1+np.exp(theta0+theta1*X[:,1]))
#cohérence: numérateur /dénominateur  avec X[:,1]
gradienttheta0=-(y[1]-1/(1+np.exp(-theta0-theta1*X[1,1])))
#gradienttheta0=-np.sum(y- functionlogistic)
gradienttheta1=-(X[1,1]*y[1]- 1/(1+np.exp(-theta0-theta1*X[1,1])))
#gradienttheta1=-np.sum(X[:,1]*y- functionlogistic)



while gradienttheta0 > 0.01 and gradienttheta1 > 0.01:
  #Attention: mettre les deux gradients dans la même boucle while car dans la formule vue en cours, les variables theta0 et theta1 sont dépendantes
  
  
  theta0=theta0-n*gradienttheta0
  #j'ai choisi arbitrairement y[1] vu que dans cette méthode on sélectionne un seul y
  gradienttheta0=-(y[1]-1/(1+np.exp(-theta0-theta1*X[1,1])))
  #attention: theta0= theta0- n*gradient ne pas oublier le gradient
  theta1=theta1-n*gradienttheta1
  #attention: theta1= theta1- n*gradient ne pas oublier le gradient
  gradienttheta1=-(X[1,1]*y[1]- 1/(1+np.exp(-theta0-theta1*X[1,1])))
  
print('Theta0: ', theta0)
print('Theta1: ', theta1)
print('Gradient pour theta0',gradienttheta0)
print('Gradient pour theta1',gradienttheta1)


#calculer la précision
iterat=0
for i in range(len(X)):
  if functionlogistic.all()<0.5:
    y[i]=0
  else:
    y[i]=1
    iterat=iterat+1
    
    

print("Nombre de y=1: ",iterat)
print("La précision est égale à:",(iterat/len(X))*100)
```

**Exercise 6 :** Same exercise that previous section but now with 3 different classes. (you have now to implement Softmax Regression instead of Logistic Regression)

```

def Softmax(vect):
    expx = np.exp(vect)
    return expx / expx.sum()

#dénominateur retourne bien un array et non un nombre!!!

#déclaration de x2 avec 3 différentes classes
x2 = np.array([[1, 4, 3, 6],  
               [2, 4, 5, 6],  
               [1, 9, 3, 6]]) 
print(x2.shape)
Softmax(x2)



```
